## Exploratory Data Analysis
### Univariate Plots

#### 1) Distribution Plots

```python
for _, feature in enumerate(list(df1.columns)[:-1]):
    fg = sns.FacetGrid(df1, hue = 'Species',height=5)
    fg.map(sns.distplot, feature).add_legend()
    plt.show()
```

#### 2) Box Plots

```python
sns.boxplot(x='Species',y='SepalLengthCm', data=df1) 
plt.show()
sns.boxplot(x='Species',y='SepalWidthCm', data=df1) 
plt.show()
sns.boxplot(x='Species',y='PetalLengthCm', data=df1) 
plt.show()
sns.boxplot(x='Species',y='PetalWidthCm', data=df1) 
plt.show()
```

Height of box represents all data between 25th percentile to 75th percentile and that horizontal bar represents maximum range of that data and width of box represents spread of that data in dataset.

#### 3) Violin Plots

```python
sns.violinplot(x="Species", y = "SepalLengthCm",data = df1,size = 8)
plt.show()
sns.violinplot(x="Species", y = "SepalWidthCm",data = df1,size = 8)
plt.show()
sns.violinplot(x="Species", y = "PetalLengthCm",data = df1,size = 8)
plt.show()
sns.violinplot(x="Species", y = "PetalWidthCm",data = df1,size = 8)
plt.show()
```

Violin plot is the combination of a box plot and probability density function.It is same as Box whiskers plot, only difference is instead of box, histogram will represent spread of data

### Bi-Variate analysis

####  Pair Plot

```python
plt.close();
sns.set_style("whitegrid");
sns.pairplot(df1.loc[:, df1.columns != 'Id'], hue="Species", height=3);
plt.show()
```

#### Observations: 
The plot between PetalLengthCm and PetalWidthCm is comparatively better. While Iris-Setosa can be easily identified Iris-Versicolor and Iris-Virginica have some overlap. We can use **“if-else”** condition (***Decision Tree***) to build a simple model to classify the flower types.

```python
from sklearn.model_selection import train_test_split  
X_train, X_test, y_train, y_test = train_test_split(x, y, 
                            test_size=0.2, random_state=0) 
```

### Training the decision tree classifier

```python
from sklearn.tree import DecisionTreeClassifier
dtree=DecisionTreeClassifier()
tree = dtree.fit(X_train,y_train)
```

```python
y_pred = dtree.predict(X_test)
```

```python

print("The parameters of the used decesion tree are:")
print(dtree.get_params)
```

Three accuracy metrics viz :
<ol>
    <li> Accuracy Score </li>
    <li> Classification Report </li>
    <li> Confusion Matrix </li>
</ol>
are presented for this Decision tree classifier.
    

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred,normalize = True)
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test,y_pred))
```

```python
from sklearn.metrics import confusion_matrix
confusion_matrix(y_test, y_pred)
```

```python
# Import necessary libraries for graph viz
from sklearn.externals.six import StringIO  
from IPython.display import Image  
from sklearn.tree import export_graphviz
import pydotplus
```

```python
dot_data = StringIO()
export_graphviz(dtree, out_file=dot_data, feature_names=iris.feature_names,  
                filled=True, rounded=True,
                special_characters=True)
graph = pydotplus.graph_from_dot_data(dot_data.getvalue())  
Image(graph.create_png())
```
